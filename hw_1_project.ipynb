```python
# Import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data
customers = pd.read_csv('Data/customers.csv')
transactions = pd.read_csv('Data/transactions.csv')
marketing_campaigns = pd.read_csv('Data/marketing_campaigns.csv')
web_sessions = pd.read_csv('Data/web_sessions.csv')
```

### Specific questions

Q1. What is the average purchase value for customers from Spain?

```python
df = transactions.merge(customers, on='customer_id', how='left')
avg_purchase_value_spain = df[df["country"] == "Spain"]["purchase_value"].mean()
print(f"Average purchase value for customers in Spain: {avg_purchase_value_spain:.2f} €")
```

Q2. What is the minimum purchase value for customers from Germany?

```python
df = transactions.merge(customers, on='customer_id')
min_purchase_value_germany = df[df["country"] == "Germany"]["purchase_value"].min()
print(f"Minimum purchase value for customers in Germany: {min_purchase_value_germany:.2f} €")
```

Q3. How much did males spend on buying products in the Sports category?

```python
df = transactions.merge(customers, on='customer_id')
sum_spent_males = df[
    (df["product_category"] == "sports") &
    (df["gender"] == "Male")
]["purchase_value"].sum()
print(f"How much males spend on buying products in the sports category: {sum_spent_males} €")
```

Q4. What is the average age of female customers from Canada?

```python
df = transactions.merge(customers, on='customer_id')
avg_age_f_canada = df[df["country"] == "Canada"]["age"].mean()
print(f"Average age of female customers from Canada: {avg_age_f_canada:.2f} years")
```

Q5. How much does an average male customer from the UK spend?

```python
df = transactions.merge(customers, on='customer_id')
sum_spent_males = df[
    (df["country"] == "UK") &
    (df["gender"] == "Male")
]["purchase_value"].mean()
print(f"How much does an average male customer from the UK spend: {sum_spent_males:.2f} €")
```

Q6. What is the maximum purchase value for females in the Toys category?

```python
df = transactions.merge(customers, on='customer_id')
max_purchaseValue_f_toys = df[
    (df["product_category"] == "toys") &
    (df["gender"] == "Female")
]["purchase_value"].max()
print(f"What is the maximum purchase value for female toys category: {max_purchaseValue_f_toys} €")    
```

Q7. What is the average number of pages viewed for customers who made a purchase (converted)?

```python
df = transactions.merge(web_sessions, on='customer_id')
avg_no_pages_viewed = df[df["converted"] == 1]["pages_viewed"].mean()
print(f"Average number of pages viewed for customers who converted: {avg_no_pages_viewed:.2f} pages")
```

Q8. What is the average time spent on the site for purchasing vs non-purchasing customers?

```python
purchasing_customers = transactions["customer_id"].unique()
customers["is_purchasing"] = customers["customer_id"].isin(purchasing_customers)
df = customers.merge(web_sessions, on='customer_id')
avg_num_pages_viewed_purchasing = df[df["is_purchasing"] == True]["time_on_site"].mean()
avg_num_pages_viewed_non_purchasing = df[df["is_purchasing"] == False]["time_on_site"].mean()
print(f"Average time on site for purchasing customers: {avg_num_pages_viewed_purchasing:.2f} \nvs Average time on site for non-purchasing customers: {avg_num_pages_viewed_non_purchasing:.2f}")
```

### Visualisation tasks

Q9. Plot purchasing value across years.

```python
plt.figure(figsize=(8,5))
transactions["purchase_date"] = pd.to_datetime(transactions["purchase_date"])
transactions["year"] = transactions["purchase_date"].dt.year

sns.lineplot(data=transactions, x="year", y="purchase_value")

plt.xlabel("Year")
plt.ylabel("Purchase value")
plt.title("Plot purchasing value across years")
#transactions["year"] = transactions["purchase_date"].dt.year
```

Q10. Plot purchasing value across product categories.

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=transactions, x="product_category", y="purchase_value")

plt.xlabel("Product category")
plt.ylabel("Purchase value")
plt.title("Plot purchasing value across product categories")
```
